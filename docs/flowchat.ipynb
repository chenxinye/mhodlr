```python
import matplotlib.pyplot as plt

import matplotlib.patches as patches



# Create figure and axis

fig, ax = plt.subplots(figsize=(10, 8))

ax.axis('off')



# Define the flowchart components and their positions

components = {

    "Input Data Preparation": (0.5, 0.9),

    "(Mixed precision) HODLR Representation Construction": (0.5, 0.7),

    "Customized Precision": (0.5, 0.5),

    "(Mixed-Precision) Matrix Operation": (0.5, 0.3),

    "Output Results": (0.5, 0.1),

}



# Draw rectangles for each component

for text, (x, y) in components.items():

    rect = patches.FancyBboxPatch(

        (x - 0.2, y - 0.05), 0.4, 0.1, boxstyle="round,pad=0.1", edgecolor="black", facecolor="#add8e6"

    )

    ax.add_patch(rect)

    ax.text(x, y, text, ha="center", va="center", fontsize=10)



# Draw arrows between components

arrows = [

    ((0.5, 0.85), (0.5, 0.75)),

    ((0.5, 0.65), (0.5, 0.55)),

    ((0.5, 0.45), (0.5, 0.35)),

    ((0.5, 0.25), (0.5, 0.15)),

]



for start, end in arrows:

    ax.annotate(

        "",

        xy=end,

        xytext=start,

        arrowprops=dict(arrowstyle="->", lw=1.5, color="black"),

    )



# Display the flowchart

plt.tight_layout()
plt.savefig('flowchart.pdf', bbox_inches='tight')
plt.show()
```
